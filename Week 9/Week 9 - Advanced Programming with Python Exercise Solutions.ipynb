1. Import peanut_lines.csv and store data as lines (encoding='ISO-8859-1')
2. Show the first 25 rows of lines

```python
import pandas as pd
import numpy as np
```

```python
lines = pd.read_csv('peanut_lines.csv', encoding='ISO-8859-1')
lines.head(25)
```

3. Using the re module, count the number of lines that have Bailey as a parent.
<br>**Hint:** Write a for loop that appends to a list based on the index of the loop**

```python
import re

bailey_count = []
for i in range(0,len(lines)):
    if re.search('Bailey',lines['Parentage'].iloc[i]):
        bailey_count.append(i)
    else:
        continue
```

```python
len(bailey_count)
```

4. Create a new column called 'Bailey_Parent' where the value is 'Yes' if 'Bailey' was used in the development of that line, otherwise use 'No'. 
5. Show the first 10 rows of lines.

```python
def bailey(x):
    if re.search('Bailey', x):
        return 'Yes'
    else:
        return 'No'
    
lines['Bailey_Parent'] = lines['Parentage'].apply(bailey)
lines.head(10)
```

6. What percentage of lines were developed using 'Bailey' are verified as high-oleic (Fatty Acid Genotype [FAG] = 'ol ol')? Round your answer to the tenth. Do not include the 'Check' lines.
    
    Note: Bailey (FAG) = '+ +', FAG = 'Check', means it hasn't been identified as high-oleic yet.

```python
bailey_perc = round(len(lines[(lines['Bailey_Parent'] == 'Yes') & (lines['FAG'] == 'ol ol')]) / len(lines) * 100, 1)
print('%s%%'%(bailey_perc))
```

7. Replace the name 'Bailey' with the tested name 'N03081T' permanently for all instances in the 'Parentage' column.
<br>**Hint:** Look up the replace() function for use on dataframes
8. Show the first 10 lines

```python
lines['Parentage'] = lines['Parentage'].str.replace('Bailey','N03081T')
lines.head(10)
```

9. Load the 'peanut_yield.txt' dataset as 'yields'
10. Merge (right join, how='Right') 'lines' and 'yields' together on 'NC_Accession' and name it 'peanut_db'

```python
yields = pd.read_csv('peanut_yield.txt', header=0, sep='\t')
peanut_db = pd.merge(lines, yields, on='NC_Accession', how='right')
```

```python
peanut_db.head()
```

11. Create a dataframe that contains the average yield of Bailey and the average of all other lines with Bailey as the parent.

```python
peanut_db = peanut_db[(peanut_db['NC_Accession'] == 'Bailey') | (peanut_db['Bailey_Parent'] == 'Yes')]
```

```python
Line = ['Bailey','Other']
Yield = [peanut_db[peanut_db['NC_Accession'] == 'Bailey']['Yield'].mean(), peanut_db[peanut_db['NC_Accession'] != 'Bailey']['Yield'].mean()]
df = pd.DataFrame(Yield, Line, columns=['Yield'])
df
```
